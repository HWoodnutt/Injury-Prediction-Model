```python
injury_prediction_dataset = injury_dataset.drop(columns=["player_index", "player_url", "player_name", "X2ndclub", "X3ndclub", "Dateofdeath"])
injury_prediction_dataset.head()
```

```python
import pandas as pd

injury_prediction_dataset = pd.get_dummies(injury_prediction_dataset, drop_first=True)
```

```python
injury_prediction_dataset.head()
```


Now that we have the variables we want to work with, we need to encode the categorical variables using OneHotEncoder.


Just like the categorical variables, we need to scale the numeric variables using MinMaxScaler.

```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

scaler = MinMaxScaler()

# Assume market_dataset is already a DataFrame with only numeric columns
num_df = pd.DataFrame(
    scaler.fit_transform(injury_prediction_dataset),
    columns=injury_prediction_dataset.columns
)

num_df.head()
```

We combine the categorical and numeric variables into X and separate our label column into Y.

```python
X = injury_prediction_dataset.drop(columns=["days_missed_current_season"])
y = injury_prediction_dataset["days_missed_current_season"]
```

We split the dataset into training and testing sets.

***1. Assigned the mean value to any null values ***

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')
X = imputer.fit_transform(X)

```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)

print ('Training set: %d rows\nTest set: %d rows' % (X_train.shape[0], X_test.shape[0]))
```

```python
X_train
```

We train our model with the training dataset.

```python
from sklearn.linear_model import LinearRegression


model = LinearRegression()

model.fit(X_train, y_train)
```

Now that we have the trained model, we can use it to make predictions on the test set.

```python
predictions = model.predict(X_test)
predictions[0:5]
```

Next, we can use the predictions on the test set to evaluate our regression model.

```python
from sklearn.metrics import mean_squared_error, r2_score

predictions = model.predict(X_test)


mse = mean_squared_error(y_test, predictions)
print("MSE:", mse)

rmse = np.sqrt(mse)
print("RMSE:", rmse)

r2 = r2_score(y_test, predictions)
print("R2:", r2)
```
